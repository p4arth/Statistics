```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# importing libraries
```

```python
data = pd.read_csv('/content/drive/MyDrive/kaggle_survey_2021_responses.csv') # loading the dataframe
```

```python
data.head() # seeing the dataframe
```

```python
dic = dict(data)
print(dic)
```

```python
new_dic = {}
for key in dic: # Making the question's as the column headings for the data.
  new_dic[dic[key][0]] = dic[key][1:]
```

```python
df = pd.DataFrame(new_dic)
df.head()
```

```python
df.shape
```

```python
null_list = [x for x in df.isna().sum()]
plt.hist(null_list)
# we can see a lot of null values in the last few questions
```
